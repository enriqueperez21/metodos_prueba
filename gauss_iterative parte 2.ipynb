## Grafique

```python
import matplotlib.pyplot as plt
x, y = zip(*xs)

new_x = []
for xi in x:
    new_x.append(float(xi))

new_y = []
for yi in y:
    new_y.append(float(yi))
fig, ax = plt.subplots(figsize=(10,7))
ax.plot(new_x, new_y, color='blue', label='Datos predecidos')
ax.legend()
```
